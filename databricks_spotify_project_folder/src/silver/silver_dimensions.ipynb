## **DimUser**

#### **AUTOLOADER**

```python
df_user = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format","parquet")\
            .option("cloudFiles.schemaLocation","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/checkpoint")\
            .option("schemaEvolutionMode","addNewColumns")\
            .load("abfss://bronze@mgrspotifyproject.dfs.core.windows.net/DimUser")
```

```python
display(df_user, checkpointLocation="abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/display_checkpoint")
```

```python
df_user = df_user.withColumn("user_name",upper(col("user_name")))
display(df_user, checkpointLocation="abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/display_checkpoint2")
```

```python
df_user_obj = reusable()

df_user = df_user_obj.dropColumns(df_user,['_rescued_data'])
df_user = df_user.dropDuplicates(['user_id'])
display(df_user, checkpointLocation="abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/display_checkpoint3")
```

```python
df_user.writeStream.format("delta")\
            .outputMode("append")\
            .option("checkpointLocation","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/checkpoint")\
            .trigger(once=True)\
            .option("path","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimUser/data")\
            .toTable("spotify_catalog.silver.dimuser")
```

## **DimTrack**

```python
df_track = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format","parquet")\
            .option("cloudFiles.schemaLocation","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimTrack/checkpoint")\
            .option("schemaEvolutionMode","addNewColumns")\
            .load("abfss://bronze@mgrspotifyproject.dfs.core.windows.net/DimTrack")
```

```python
display(df_track, checkpointLocation="abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimTrack/display_checkpoint")
```

```python
df_track = df_track.withColumn("durationFlag",when(col('duration_sec')<150,"low")\
                                            .when(col('duration_sec')<300,"medium")\
                                            .otherwise("high"))

df_track = df_track.withColumn("track_name",regexp_replace(col('track_name'),'-',' '))

df_track = reusable().dropColumns(df_track,['_rescued_data'])

display(df_track, checkpointLocation="abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimTrack/display_checkpoint2")
```

```python
df_track.writeStream.format("delta")\
            .outputMode("append")\
            .option("checkpointLocation","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimTrack/chekpoint")\
            .trigger(once=True)\
            .option("path","abfss://silver@mgrspotifyproject.dfs.core.windows.net/DimTrack/data")\
            .toTable("spotify_catalog.silver.dimtrack")
```
